```python
raw_df = tobacco.copy() 

columns_to_keep = [
    'State', 'Area',

    # Smoking behavior
    'Ever_Tob_Use', 'Curr_Tob_Use', 'Ever_Cig', 'Curr_Cig', 'Age_Init_Cig','Curr_Smoke',

    # Exposure
    'Smoke_Home', 'Smoke_Enclosed', 'Smoke_Outdoor', 'Seen_Smoke_School', 'Smoke_Exposure',

    # Media & messaging
    'Seen_AT_Message', 'Seen_AT_Media', 'Seen_AT_Events', 'Seen_Warnings', 
    'Seen_Ads', 'Seen_Use_Media', 'Seen_Cig_Ads_POS',

    # Awareness
    'Taught_Tob_Effects', 'Aware_COTPA', 'Aware_Board_Policy', 'Tobacco_Free_Guideline', 'Fine_Auth',

    # Quit struggle
    'TryQuit_Smoke_12mo', 'WantQuit_Smoke', 'Quit_Smoke_12mo', 'Hard_To_Quit',

    # Access
    'Source_Cig_Store', 'Source_Cig_Paan', 'Bought_Cig_Loc', 'Refused_Cig_Sale',
]

df = tobacco[columns_to_keep]
df.head()
```

```python
df['Area'] = df['Area'].str.strip().str.title()

df = df[df['Area'] == 'Total'].reset_index(drop=True)
df.drop('Area', axis=1)
```

Question 1: Does Exposure at School or Home Increase Tobacco Usage?Â¶
        Assumption
        Youth who witness smoking behaviorâ€”either at home or within the school environmentâ€”are more likely to start using tobacco themselves. Peer influence, normalization, and lack of adult control may accelerate experimentation and adoption.

        This section investigates:

        If students who report seeing smoking at school show higher tobacco use
        Whether home exposure correlates with earlier initiation or increased likelihood of trying tobacco

```python
corr_q1 = df[['Seen_Smoke_School', 'Ever_Tob_Use','Age_Init_Cig','Smoke_Home']].corr()
plt.title("Correlation between Seen smoking in school and early usage of smoke.")


heatmap = sns.heatmap(
    corr_q1, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm", 
    linewidths=0.5, 
    linecolor='gray',
    cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'}
)

plt.title("Correlation Between Smoking Exposure at School and Tobacco Use", fontsize=14, weight='bold', pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
```

Conclusion: Influence of Smoking Exposure on Youth Tobacco Use
The correlation analysis provides strong support for the hypothesis that youth who are exposed to smokingâ€”either at home or within school environmentsâ€”are more likely to experiment with or regularly use tobacco products.

Home exposure to smoking demonstrates the strongest correlation with tobacco use (r = 0.81), followed closely by its correlation with school exposure (r = 0.78). This suggests that permissive environments in both private and public settings may reinforce each other, compounding the risk of tobacco initiation.

Exposure to smoking in schools also shows a high correlation with tobacco use (r = 0.75), underscoring the influence of peer behavior and the potential lack of enforcement in institutional settings.

Age of cigarette initiation exhibits weak or negative correlations with exposure factors (r â‰ˆ -0.03 to -0.10), indicating that while early exposure may not strongly determine how young individuals begin smoking, it significantly increases the likelihood that they will try tobacco at some point.

These findings highlight the critical role that social and environmental factors play in youth tobacco use. Both home and school serve not only as physical spaces but as behavioral environments that can normalize harmful habits. Question 2: How Does Media Exposure Influence Tobacco Usage?
        Assumption
        Students who frequently encounter tobacco use in films, online content, or advertising may become desensitized or more curious about trying it. Conversely, exposure to anti-smoking campaigns should discourage usage by raising awareness of health risks.

        This section explores:

        Whether students who see tobacco use in media are more likely to report usage
        Whether anti-tobacco messaging correlates with lower current tobacco consumption
        Potential contradictions between prevention campaigns and glamorized portrayals

```python

corr_q2 = df[['Seen_Use_Media', 'Curr_Tob_Use', 'Ever_Tob_Use', 'TryQuit_Smoke_12mo']].corr()


plt.title("Correlation between Smoking in youth and watching smoking in Media")


heatmap = sns.heatmap(
    corr_q2, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm", 
    linewidths=0.5, 
    linecolor='gray',
    cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'}
)

plt.title("Correlation Between Smoking Exposure at Media Use", fontsize=14, weight='bold', pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()




plt.show()
```

Conclusion: Media Influence and the Mixed Signals on Tobacco Use
The analysis reveals a nuanced relationship between media exposure and tobacco usage among students, aligning partially with the initial assumptions.

Exposure to tobacco use in media (e.g., films, online platforms, advertisements) shows a moderate correlation with attempts to quit smoking (r = 0.32) and a weak positive correlation with both current (r = 0.16) and ever tobacco use (r = 0.07). This suggests that media may have a dual influenceâ€”normalizing tobacco behavior while also prompting awareness or contemplation of quitting when paired with anti-smoking messaging.

Current and ever tobacco use remain strongly correlated with each other (r = 0.94), indicating consistent behavior once initiation occurs. However, the weak correlation between media exposure and initiation suggests that media alone may not be the primary driver, but a reinforcing factor in the broader behavioral ecosystem.

Interestingly, the correlation between media exposure and quit attempts (r = 0.32) may imply increased awareness or guilt stemming from health campaigns embedded within media or from internal conflict created by contradictory portrayals.

Implications
These findings underscore the complexity of media influence. While glamorized portrayals of tobacco may subtly encourage curiosity, concurrent exposure to anti-tobacco messaging can drive quit attempts. Therefore, media literacy education, alongside targeted public health messaging, may help students critically process tobacco content, reducing the risk of harmful behavior uptake.
